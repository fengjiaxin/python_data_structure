```python
from BubbleSort import BubbleSortEasy,BubbleSortImp
from InsertSort import InsertSort
from MergeSort import mergeSort
from QuickSort import QuickSort
from SelectionSort import SelectionSort
```

```python
import time
```

```python
def test(nums,SortType):
    print('排序之前：')
    print(str(nums))
    start = time.time()
    SortType(nums)
    t = time.time() - start
    print('===========================')
    print('排序之后：')
    print(str(nums))
    print('===========================')
    print('排序运行时间：'+str(t) + ' s')    
```

```python
nums = list(range(100))
```

```python
str(nums)
```

### 打乱列表顺序

```python
import random
```

```python
random.shuffle(nums)
```

### 测试冒泡排序

```python
test(nums.copy()  ,BubbleSortEasy)
```

```python
test(nums.copy()  ,BubbleSortImp)
```

### 测试插入排序

```python
test(nums.copy()  ,InsertSort)
```

### 测试选择排序

```python
test(nums.copy()  ,SelectionSort)
```

### 测试快速排序

```python
test(nums.copy()  ,QuickSort)
```

### 测试归并排序

```python
test(nums.copy()  ,mergeSort)
```
